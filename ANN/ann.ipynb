importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
tf.__version__
```

reading the dataset

```python
dataset=pd.read_csv('Churn_Modelling.csv')
dataset
```

```python
dataset.info()
```

```python
X=dataset.iloc[:,3:-1].values
y=dataset.iloc[:,-1].values
```

```python
print(X)
print(X.shape)
```

```python
print(y)
print(y.shape)
```

## Preprocessing

label encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,2]=le.fit_transform(X[:,2])
print(X)
```

one hot encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
print(X)
```

train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

feature scaling (compulsory for deep learning and applied for all features)

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
print(X_train)
print(X_test)
```

## Build ANN

init the ann

```python
ann=tf.keras.models.Sequential()
```

adding the input(automatically initialised with neurons = no of features) layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

adding the second layer

```python
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

adding the output layer

```python
#if binary classification then sigmoid
#else softmax
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

## Training the ANN

Comiling the ANN

```python
#adam uses stochastic gradient descent to minimize the cost function
#if binary classification(i.e. o/p is 0 or 1) then use binary_crossentropy
#if non-binary classification then use categorical_crossentropy
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

Training the ANN

```python
ann.fit(X_train, y_train, batch_size=32, epochs=1000)
```
